# Lógica de Programação - Passo a passo para a solução do desafio

```python
# Passo 0 - Entender o Desafio que vai resolver
# Passo 1 - Importar nossa base de dados, planilha (pasta vendas)
import pandas as pd
import plotly.express as px
import smtplib
import email.message

tabela_vendas = pd.read_excel("/content/drive/MyDrive/Pyhon para data science/Vendas/Vendas+-+Base+de+Dados.xlsx")
display(tabela_vendas) # - Uso o display para mostrar a planilha de maneira organizada, como uma verificação positiva da importação da planilha
```

```python
# Passo 2 - Calcular o produto mais vendido em quantidade

tb_qtd_produtos = tabela_vendas.groupby('Produto').sum()
tb_qtd_produtos = tb_qtd_produtos[['Quantidade']]
tb_qtd_produtos = tb_qtd_produtos.sort_values(by='Quantidade', ascending=False)
display(tb_qtd_produtos)
```

```python
# Passo 3 - Qualcular o produto mais vendido em faturamento
tabela_vendas['Faturamento'] = tabela_vendas['Quantidade'] * tabela_vendas['Valor Unitário']
tb_produto_mais_vendido = tabela_vendas.groupby('Produto').sum()
tb_produto_mais_vendido = tb_produto_mais_vendido[['Faturamento']]
tb_produto_mais_vendido = tb_produto_mais_vendido.sort_values(by='Faturamento', ascending=False)
display(tb_produto_mais_vendido)
```

```python
# Passo 4 - Calcular Loja/Estado que mais vendeu (faturamento)]
tb_fatur_loja = tabela_vendas.groupby('Loja').sum()
tb_fatur_loja = tb_fatur_loja[['Faturamento', 'Quantidade']].sort_values(by='Faturamento', ascending=False)
print(tb_fatur_loja)
```

```python
# Passo 5 - Calcular o ticket médio por loja/estado
tabela_vendas['Ticket_Médio'] = tabela_vendas['Valor Unitário']
tb_ticket_medio = tabela_vendas.groupby('Loja').mean(numeric_only=True)
tb_ticket_medio = tb_ticket_medio[['Ticket_Médio']].sort_values(by='Ticket_Médio', ascending=False)
display(tb_ticket_medio)
```

```python
# Passo 6 - Criar um dashboard da loja que mais vendeu em Faturamento
grafico = px.bar(tb_qtd_produtos, y='Quantidade', x=tb_qtd_produtos.index)
grafico.show()
```
